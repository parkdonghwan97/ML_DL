```python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df = pd.read_csv('data/test.csv', header=None)
df.columns=['q1','q2','mid','final']
df.head(3)
```

```python
# Matmul 사요해야하지때문에 q1 q2 mid 를 n행 3열 매트릭스로

x_data = df.iloc[:,:-1].values
y_data = df.iloc[:,[-1]].values
```

```python
#  df.iloc[:,:-1].values
```

```python
x_data.shape, y_data.shape
```

```python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
```

```python
''' 
3행 1열     
q1
w2
mid
'''
w = tf.Variable(tf.random.uniform(  [3,1] )   ) # 특성데이터갯수, label개수
b = tf.Variable(tf.random.uniform([1]) )   # [라벨개수]
```

```python
def compute_loss():
    hx = tf.matmul(x,w) + b
    cost = tf.reduce_mean( (hx-y)**2  )
    return cost
```

```python
optimizer = Adam(0.1)
for i in range(1000):
    optimizer.minimize(compute_loss,var_list=[w,b] )
    print(i,'cost: ',compute_loss().numpy())
```

```python
def hxfn(xdata):
    xd = tf.constant(xdata , tf.float32)
    hx = tf.matmul(xd,w) + b 
    return hx.numpy()
```
