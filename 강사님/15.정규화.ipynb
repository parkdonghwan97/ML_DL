```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler, StandardScaler

from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
data = [[828, 920, 1234567, 1020, 1111],
            [824, 910, 2345612, 1090, 1234],
            [880, 900, 3456123, 1010, 1000],
            [870, 990, 2312123, 1001, 1122],
            [860, 980, 3223123, 1008, 1133],
            [850, 970, 2432123, 1100, 1221]]
data = np.float32(data)
```

```python
df = pd.DataFrame( data )
df
```

```python
x_data = data[:,:-1]
y_data = data[:, [-1] ]
```

```python
x_data
```

```python
scaleF = MinMaxScaler()
x_dataN = scaleF.fit_transform( x_data) #matrix
```

```python
x_dataN
```

```python
scaleL = MinMaxScaler()
y_dataN = scaleL.fit_transform( y_data) #matrix
```

```python
y_dataN
```

```python
dense = Dense( units=1, input_dim=4)
model = Sequential( [dense])
model.compile( loss='mse', optimizer = Adam( 0.1) )
h = model.fit(x_dataN, y_dataN, epochs=1000)
```

```python
h.history.keys()
```

```python
plt.plot( h.history['loss'] )
plt.show()
```

```python
w,b = model.get_weights()
w
```
