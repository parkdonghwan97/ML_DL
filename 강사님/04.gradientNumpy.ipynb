```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def cost(x, y, w):
    hx = w*x # [1,2,3] X -1  => hx = [-1,-2,-3]
    #= [-1,-2,-3] - [1,2,3] =[-2,-4,-6]**2    
    c = np.sum( (hx - y)**2 )/len(x) 
    return c
```

```python
def gradient( x, y , w):
    c = np.sum( (w*x-y)*x*2  )
    return c/len(x)
```

```python
x_data = np.array( [1,2,3] )
y_data = np.array( [1,2,3] )
```

```python
cost( x_data, y_data, -1)
```

```python
w = 10
for epoch in range(200):
    c = cost( x_data, y_data, w)
    g = gradient( x_data,y_data, w)
    w = w-0.1*g
    print( epoch, "cost=", c, "w=", w)
print("w=", w)
```
