# 보스턴 데이터 셋 정규화 후 합습 하여 첫째 행 medv 예측

```python
from sklearn.datasets import load_boston
boston = load_boston()
boston.keys()

df_boston =pd.DataFrame(boston['data'],columns=boston['feature_names'])
df_boston['medv'] = boston['target']
x = df_boston.iloc[ :, : -1]
y = boston['target'] # y = df_boston[:, [-1]]
df_boston.head(3)
```

```python
x
```

```python
y.shape
```

```python
y_data = y.reshape(-1,1)
```

```python
scalerF = MinMaxScaler()
x_dataN = scalerF.fit_transform(x)  
```

```python
scalerL = MinMaxScaler()
y_dataN = scalerL.fit_transform(y_data) 
```

```python
x_dataN.shape
```

```python
dense = Dense(units =1 , input_dim = 13 )
model = Sequential([dense])
model.compile(loss='mse',optimizer=Adam(0.1))
h = model.fit(x_dataN, y_dataN ,epochs=1000)
```
