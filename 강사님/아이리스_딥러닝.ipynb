```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
iris
```

```python
irisDF=pd.DataFrame(iris['data'])
irisDF.columns=iris['feature_names']
irisDF['species'] = iris['target']
irisDF.head()
```

```python
x_data = irisDF.iloc[:,:-1].values
x_data
```

```python
y_data=pd.get_dummies(irisDF['species']).values
y_data
# 0 -> 1, 0, 0
# 1 -> 0, 1, 0
# 2 -> 0, 0, 1
```

```python
x = tf.constant( x_data, tf.float32)
y = tf.constant( y_data, tf.float32)
```

```python
type(x_data)
```

```python
initX=tf.initializers.GlorotUniform()


w1=tf.Variable(initX(shape=[4,64]))
b1=tf.Variable(initX(shape=[64]))

w2=tf.Variable(initX(shape=[64,3]))
b2=tf.Variable(initX(shape=[3]))
```

```python
def compute_loss():
    z1=tf.matmul(x,w1)+b1
    hx1=tf.nn.softmax(z1)
    
    z=tf.matmul(hx1, w2)+b2
    cost_i = tf.nn.softmax_cross_entropy_with_logits( labels=y, logits=z )
    cost = tf.reduce_mean( cost_i)
    return cost
```

```python
optimizer=Adam(learning_rate=0.1)
for i in range(2000):
    optimizer.minimize(compute_loss, var_list=[w1,w2, b1, b2])
    print(i, compute_loss().numpy())
```

```python
def hxfn(xdata):
    xd=tf.constant(xdata, tf.float32)
    z1=tf.matmul(xd,w1)+b1
    hx1=tf.nn.softmax(z1)
    z=tf.matmul(hx1,w2)+b2
    hx = tf.nn.softmax( z )
    return (hx.numpy()>0.5)+0
```

```python
pred=[[5.1, 3.5, 1.4, 0.2]]
```

```python
hxfn(pred)
```

```python
hxfn(x_data)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score
accuracy = accuracy_score(y_data, hxfn(x_data))
accuracy
```

```python
dense1 = Dense( units=64, input_dim=4 ,activation="softmax") # w:4x64, b:64
dense2 = Dense( units=3, input_dim=64 ,activation="softmax") # w:64x1, b:3
model = Sequential( [dense1, dense2])
model.compile( loss="categorical_crossentropy" ,optimizer=Adam(0.1), metrics=['acc'] )
h = model.fit( x_data, y_data, epochs=1000)
```
