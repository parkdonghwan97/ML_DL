```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential
from tensorflow.keras.optimizers import Adam
import pandas as pd
```

```python
df = pd.read_csv( 'data/softmax.txt', sep=' ',header=None)
df
```

```python
x_data = df.iloc[:,0:2].values
y_data = df.iloc[:,2:].values
```

```python
dense = Dense( units=3, input_dim=2 ,activation="softmax") # w:2x3, b:3
model = Sequential( [dense])
model.compile( loss="categorical_crossentropy" ,optimizer=Adam(0.1), metrics=['acc'] )
h = model.fit( x_data, y_data, epochs=1000)
```

```python
def resultFn( dt ):
    n = int(  model.predict( dt ).argmax( axis=1) )
    grade = ['A','B','C']
    return grade[n]
```
