```python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/cars.csv', index_col='Unnamed: 0')
df
```

```python
x_data = df['speed'].values
y_data = df['dist'].values
```

```python
plt.scatter( x_data, y_data)
plt.show()
```

```python
x = tf.constant( x_data, tf.float32)
y = tf.constant( y_data, tf.float32)
```

```python
w = tf.Variable( tf.random.uniform([1]) )
b = tf.Variable( tf.random.uniform([1]) )
```

```python
def compute_cost():
    hx = w*x + b
    cost = tf.reduce_mean( (hx-y)**2)
    return cost
```

```python
optimizer = Adam( learning_rate=0.01)
hist_cost=[]
for i in range(2000):
    optimizer.minimize( compute_cost, var_list=[w,b]  )
    c = compute_cost().numpy()
    hist_cost.append( c )
    print(i, 'cost=',c )
```

```python
def hxFn( xdata):
#     xd = tf.constant( xdata)
#     hx = w*xd + b
    hx = w*xdata + b
    return hx.numpy()
```

```python
plt.plot( hist_cost)
plt.show()
```
