```python
import torch
from torch.optim import Adam
```

```python
x_data = torch.FloatTensor([1,2,3,4,5])
y_data = torch.FloatTensor([2,4,6,8,10])
```

```python
w = torch.tensor(10.0, requires_grad=True)
b = torch.tensor(10.0, requires_grad=True)
```

```python
optimizer = Adam( [w,b],lr=0.1)   
for i in range(2000):
    hx = w * x_data + b 
    cost = torch.mean( (hx -y_data)**2 )
    optimizer.zero_grad() # 기울기 초기화
    cost.backward() # cost값 미분
    optimizer.step() # w =  w-0.1 x w미분값, b= b-0.1 x b미분값
    print(i,'cost: ',cost.item())
```
