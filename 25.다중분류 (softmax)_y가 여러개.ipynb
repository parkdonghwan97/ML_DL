```python
import tensorflow as tf
import numpy as np
import pandas as pd

import math
import matplotlib.pyplot as plt 

plt.style.use('ggplot')

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('tdata/softmax.txt', sep =' ' , header = None)
df
```

```python
x_data = df.iloc[:,0:2]
y_data = df.iloc[:,2:]
```

```python
x_data.shape , y_data.shape
```

```python
x = tf.constant( x_data, tf.float32)
y = tf.constant( y_data, tf.float32)
```

```python
initX =tf.initializers.GlorotUniform()
w = tf.Variable( initX( shape=[2,3] ) ) #[특성데이터의 갯수, 라벳의갯수]
b = tf.Variable( initX( shape=[3] ) ) #라벨의 갯수
```

```python
def compute_loss():
    z = tf.matmul( x, w) + b
#     hx = tf.nn.softmax( z )
    cost_i = tf.nn.softmax_cross_entropy_with_logits( labels=y, logits=z )
    cost = tf.reduce_mean( cost_i)
    return cost
```

```python
optimizer = Adam( 0.1 )
for i in range(1000):
    optimizer.minimize( compute_loss, var_list=[w,b] )
    print( i, 'cost:', compute_loss().numpy() )
```

# 케라스 버전

```python
df = pd.read_csv('tdata/softmax.txt', sep =' ' , header = None)
df
```

```python
# x값 2  , y깂 3
x_data = df.iloc[:,0:2].values
y_data = df.iloc[:,2:].values
```

```python
x_data.shape, y_data.shape
```

```python
dense = Dense( units=3, input_dim=2 ,activation="softmax")
model = Sequential( [dense])
model.compile( loss="categorical_crossentropy" ,optimizer=Adam(0.1), metrics=['acc'] )
h = model.fit( x_data,y_data, epochs=1000)
```

```python
w, b = dense.get_weights()
w
```
