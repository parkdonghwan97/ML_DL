```python
import tensorflow as tf
import numpy as np
import math
import matplotlib.pyplot as plt 
from  tensorflow.keras.optimizers import Adam
plt.style.use('ggplot')
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
#공부한시간, 출석일수
x_data = np.array( [[1,3],[2,2],[3,1],[4,6],[5,5],[6,4]])
y_data = np.array( [[0],[0],[0],[1],[1],[1]])
```

```python
dense = Dense(units= 1, input_dim = 2 ,activation='sigmoid') # w : 2x1  b :1 
model = Sequential([dense])

# 로지스틱에서는 cost가 mse가 아님 
# model.compile(loss = 'mse')

model.compile(loss = 'binary_crossentropy',optimizer=Adam(0.1),metrics=['acc']  )
h = model.fit(x_data,y_data,epochs=1000)

```

# 파이프라인

```python
import pandas as pd
df = pd.read_csv('ddata/pima-indians-diabetes.data.csv',encoding='euc-kr')
for c in ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']:
    idx = np.where(df[c]==0)[0]
    df.loc[idx,c] = df[c].mean()
df.head(3)
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.pipeline import make_pipeline
```

```python
x_data = df.iloc[:,:-1].values
y_data = df.iloc[:,[-1]].values
```

```python
def createModel():
    dense = Dense( units=1, input_dim=8 ,activation="sigmoid") # w:2x1, b:1
    model = Sequential( [dense])
    model.compile( loss="binary_crossentropy" ,optimizer=Adam(0.1), metrics=['acc'] )

    return model
#h = model.fit( x_data, y_data, epochs=1000)
```

```python
kcf = KerasClassifier(build_fn=createModel, epochs=1000 ,verbose=1)
pmodel = make_pipeline(MinMaxScaler(),kcf)
pmodel.fit(x_data,y_data)
```

```python
pmodel.predict(x_data)
```
