```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.optimizers import SGD
```

```python
x_data = np.array([1,2,3,4,5])
y_data = np.array([3,5,7,9,11])
```

```python
x = tf.constant(x_data,tf.float32)
y = tf.constant(y_data,tf.float32)

w = tf.Variable(10.0)
b = tf.Variable(10.0)
```

```python
def compute_cost():
    hx = w*x  + b
    cost = tf.reduce_mean( (hx - y)**2 )
    return cost  
```

```python
optimizer = SGD( learning_rate=0.01 ) # 미분, 알파 기능
for i in range(2000):
    optimizer.minimize( compute_cost, var_list=[w,b] )
    '''
    w = w  - 0.01 * (w에 대한 미분값)  w.assign_sub(0.01* 미분값)
    b = b -  0.01 * (b에 대한 미분값)  b.assign_sub(0.01* 미분값)
    '''
    print(i,"코스트",compute_cost().numpy())
```
