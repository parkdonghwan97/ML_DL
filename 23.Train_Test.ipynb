```python
import tensorflow as tf
import numpy as np
import pandas as pd

import math
import matplotlib.pyplot as plt 

plt.style.use('ggplot')

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('ddata/pima-indians-diabetes.data.csv',encoding='euc-kr')
for c in ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']:
    idx = np.where(df[c]==0)[0]
    df.loc[idx,c] = df[c].mean()
df.head(3)
```

```python
x_data = df.iloc[:,:-1].values
y_data = df.iloc[:,[-1]].values
```

```python
x_train,x_test, y_train,y_test = train_test_split(x_data,y_data,test_size=0.3, stratify=y_data)
```

```python
x_train.shape,y_train.shape
```

```python
x_test.shape, y_test.shape
```

```python
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x_train)
```

```python
dense = Dense( units=1, input_dim=8 ,activation="sigmoid") # w:2x1, b:1
model = Sequential( [dense])
model.compile( loss="binary_crossentropy" ,optimizer=Adam(0.1), metrics=['acc'] )
h = model.fit( x_scaled,y_train, epochs=1000)
```

```python
# 모델 평가  evaluate(스케일x , 훈련 y)
# 훈련 데이터 정확도
model.evaluate(x_scaled, y_train)
```

```python
scaler.transform(x_test)
```

```python
# 테스트 셋에 대한 정확도
model.evaluate(scaler.transform(x_test), y_test)
```

```python
df = pd.read_csv('ddata/pima-indians-diabetes.data.csv')

x_data = df.iloc[:,:-1].values
y_data = df.iloc[:,[-1]].values

scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x_data)
```

```python
x_train,x_test,y_train, y_test = train_test_split( x_scaled, y_data, test_size=0.3, 
                                                  stratify=y_data )
```

```python
# 검증 셋 확인
x_train,x_valid,y_train, y_valid = train_test_split( x_train, y_train, test_size=0.3, 
                                                  stratify=y_train )
```

```python
dense = Dense( units=1, input_dim=8 ,activation="sigmoid") # w:2x1, b:1
model = Sequential( [dense])
model.compile( loss="binary_crossentropy" ,optimizer=Adam(0.1), metrics=['acc'] )
h = model.fit( x_train, y_train, epochs=1000,validation_data=(x_valid,y_valid))
```

```python
model.evaluate( x_train, y_train)

model.evaluate(  x_test, y_test)
```

```python
h.history.keys()
```

```python
plt.plot(h.history['loss'])
plt.plot(h.history['val_loss'])
plt.show()
```

```python
plt.plot(h.history['acc'])
plt.plot(h.history['val_acc'])
plt.show()
```
