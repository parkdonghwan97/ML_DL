# tensor cost

````python
x_data = [1,2,3]
y_data = [1,2,3]
x = tf.constant( x_data ,dtype=float)
y = tf.constant( y_data ,dtype=float)
````

````python
w = 3 
hx = w * x  # [3,6,9]
sq = tf.square( hx - y )   # [3,6,9] - [1,2,3] = [2,4,6]^2 = [4,16,36]
cost = tf.reduce_mean( sq )
````

# tensorGradient  (기울기)

````python
from tensorflow.keras.optimizers import SGD
````

````python
x_data = np.array([1,2,3,4,5])
y_data = np.array([3,5,7,9,11])
````

````python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
````

````python
x = tf.constant(x_data,tf.float32)
y = tf.constant(y_data,tf.float32)

w = tf.Variable(10.0)
b = tf.Variable(10.0)
````

````python
x.numpy(),y.numpy(),w.numpy(),b.numpy()
````

````python
# hx설정 및 cost 구하는 함수

def compute_cost():
    hx = w * x + b    # hx = wx + b
    cost = tf.reduce_mean( (hx - y)**2) #cost = (예측값 - 실제값)의 제곱
    return cost
````

경사 하강법 구현

````python
optimizer = SGD(learning_rate=0.01)   # 미분과 알파 
for i in range(2000):
    optimizer.minimize(compute_cost,var_list=[w,b] )# 손실함수의 최소값 찾기
    '''
    w = w  - 0.01 * (w에 대한 미분값)  w.assign_sub(0.01* 미분값)
    b = b -  0.01 * (b에 대한 미분값)  b.assign_sub(0.01* 미분값)
    '''
    print(i,"코스트",compute_cost().numpy())
````

# PyTorch 선형회귀

````python
import torch
from torch.optim import Adam
````

````python
x_data = torch.FloatTensor([1,2,3,4,5])
y_data = torch.FloatTensor([2,4,6,8,10])
w = torch.tensor(10.0, requires_grad=True)
b = torch.tensor(10.0, requires_grad=True)
````

경사 하강법 구현

````python
optimizer = Adam( [w,b],lr=0.1)   
for i in range(2000):
    hx = w * x_data + b 
    cost = torch.mean( (hx -y_data)**2 )
    optimizer.zero_grad() # 기울기 초기화
    cost.backward() # cost값 미분
    optimizer.step() # w =  w-0.1 x w미분값, b= b-0.1 x b미분값
    print(i,'cost: ',cost.item())  #손실 계산
````

- 내가 다중이랑 다항이랑 헷갈리는데
- 다항회귀는 독립변수의 차수를 높임. (x^2)
- 다중회귀는 다중의 독립변수가 있는 형태 (x1, x2)

````python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
````

## x1 : 공부할시간 + x2 : 출석일수 y : 점수

````python
x1_data = [1,0,3,4,5]  #공부한시간
x2_data = [0,1,2,3,4] #출석일수
y_data = [1,2,3,4,5] #점수
````

````python
x1 = tf.constant( x1_data , tf.float32)
x2 = tf.constant( x2_data , tf.float32)
y = tf.constant( y_data , tf.float32)
````

````python
w1 = tf.Variable(tf.random.uniform([1])) # x1 의 가중치
w2 = tf.Variable(tf.random.uniform([1])) # x2 의 가중치
b = tf.Variable(tf.random.uniform([1])) 
````

````python
# cost 구하는 함수
def compute_loss():
    hx = w1*x1 + w2*x2 + b
    cost = tf.reduce_mean( (hx - y)**2 )
    return cost
````

````python

optimizer = Adam(0.1)
for i in range(1000):
    optimizer.minimize(compute_loss, var_list=[w1,w2,b] ) # 미분
    print(i,'cost: ',compute_loss().numpy())
````
