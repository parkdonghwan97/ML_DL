- 내가 다중이랑 다항이랑 헷갈리는데
- 다항회귀는 독립변수의 차수를 높임. (x^2)
- 다중회귀는 다중의 독립변수가 있는 형태 (x1, x2)

````python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
plt.style.use('ggplot')
````

## x1 : 공부할시간 + x2 : 출석일수 y : 점수

````python
x1_data = [1,0,3,4,5]  #공부한시간
x2_data = [0,1,2,3,4] #출석일수
y_data = [1,2,3,4,5] #점수
````

````python
x1 = tf.constant( x1_data , tf.float32)
x2 = tf.constant( x2_data , tf.float32)
y = tf.constant( y_data , tf.float32)
````

````python
w1 = tf.Variable(tf.random.uniform([1])) # x1 의 가중치
w2 = tf.Variable(tf.random.uniform([1])) # x2 의 가중치
b = tf.Variable(tf.random.uniform([1])) 
````

````python
# cost 구하는 함수
def compute_loss():
    hx = w1*x1 + w2*x2 + b
    cost = tf.reduce_mean( (hx - y)**2 )
    return cost
````

````python

optimizer = Adam(0.1)
for i in range(1000):
    optimizer.minimize(compute_loss, var_list=[w1,w2,b] ) # 미분
    print(i,'cost: ',compute_loss().numpy())
````

````python
w1.numpy(),w2.numpy()
````

````python
# 5시간 공부하고 5시간 출석한 학생의 점수는?
# w1*5 + w2*5 + b

w1.numpy()[0]*5 + w2.numpy()[0]*5 + b.numpy()[0]
````

# 다중선형회귀(MatMul 이용)

````python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
plt.style.use('ggplot')
````

````python
df = pd.read_csv('../data/test.csv', header=None)
df.columns=['q1','q2','mid','final']
df.head(3)
````

Matmul 사용해야하지때문에 q1 q2 mid 를 n행 3열 매트릭스로


````python
df.iloc[:,:-1].values
````

````python
x_data = df.iloc[:,:-1].values
y_data = df.iloc[:,[-1]].values
````

````python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
````

````python
''' 
3행 1열     
q1
w2
mid
'''
w = tf.Variable(tf.random.uniform(  [3,1] )   ) # 특성데이터갯수, label개수
b = tf.Variable(tf.random.uniform([1]) )   # [라벨개수]
````

````python
def compute_loss():
    hx = tf.matmul(x,w) + b
    cost = tf.reduce_mean( (hx-y)**2  )
    return cost
````

````python
optimizer = Adam(0.1)
for i in range(1000):
    optimizer.minimize(compute_loss,var_list=[w,b] )
    print(i,'cost: ',compute_loss().numpy())
````

````python
w.numpy()  # q1 q2 mid 의 가중치
````

````python
b.numpy()  # 편차
````

````python
def hxfn(xdata):
    xd = tf.constant(xdata , tf.float32)
    hx = tf.matmul(xd,w) + b 
    return hx.numpy()
````

q1 q2 mid가 80 90 90 인 경우의 최종 점수 예측 

- matmul이 (x와 w) 의 행렬곱

````python
w.numpy()  #   w가  3행 1열일  => 따라서  x는 1행 3열
````

````python
hxfn([    
        [80,90,90 ]  
] )
````

# 다중선형회귀( 케라스 )

````python
import tensorflow as tf
import numpy as np
import pandas as pd

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
````

````python
df = pd.read_csv('../data/test.csv', header=None)
df.columns=['q1','q2','mid','final']
df.head(3)
````

````python
x_data = df.iloc[:,:-1].values #특성데이터 추출
y_data = df.iloc[:, [-1]].values

dense = Dense(units = 1, input_dim =3 ) # w : 3x1   b : 1   초기값
model = Sequential([dense])
model.compile(loss='mse',optimizer=Adam(0.1))
# n행 3열
h = model.fit(x_data, y_data, epochs=500) # x값을 행렬로 주어야함
````

````python
w,b = dense.get_weights()
w.shape,b.shape  # w = 3행 1열
````

````python
# q1 q2 mid  80 90 90 인 사람의 최종점수
model.predict([ 
    [80,90,90]
])
````
