```python
import tensorflow as tf
import numpy as np
import pandas as pd

import math
import matplotlib.pyplot as plt 

plt.style.use('ggplot')

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense ,Flatten
from tensorflow.keras.optimizers import Adam

from sklearn.preprocessing import MinMaxScaler,StandardScaler
from sklearn.model_selection import train_test_split


from sklearn.datasets import load_iris
# iris[5.1, 3.5, 1.4, 0.2]
```

```python
iris = load_iris()
df = pd.DataFrame(iris['data'],columns=iris['feature_names'])
df['species'] = iris['target']
df.head(3)
```

```python
# 원핫 인코딩 
temp = pd.DataFrame(data= pd.get_dummies(df['species']).values)
temp.shape , df.shape
```

```python
df = pd.concat([df,temp],  axis=1)
del(df['species'])
```

```python
df.sample(4)
```


# 1. [5.1, 3.5, 1.4,  0.2   ]   종분류 예측

```python
x_data = df.iloc[:,0:4]
y_data = df.iloc[:,4:]

x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
```

## 텐서


```python
initX = tf.initializers.GlorotUniform()


w1 = tf.Variable( initX( shape = [4,64]  )    )
b1 = tf.Variable( initX( shape=[64] ) )

w2 = tf.Variable( initX( shape=[64,32] ) )
b2 = tf.Variable( initX( shape=[32] ) )

w3 = tf.Variable( initX(shape=[32,3]))
b3 = tf.Variable( initX(shape=[3]))
```

```python
def compute_loss():
    z1 = tf.matmul(x,w1) + b1  # x 와 w1 matmul  4행2열 X 2행 2열 => 4행 2열 
    hx1 = tf.nn.softmax( z1 )  # z1을 sigmoid 함수에 넣음
    
    z2 = tf.matmul( hx1, w2  ) + b2  # 
    hx2 = tf.nn.softmax( z2 )
    
    z = tf.matmul(hx2 , w3 ) + b3
#     hx = tf.nn.sigmoid( z )   
    
    cost_i =tf.nn.softmax_cross_entropy_with_logits(logits=z,labels =y )
    cost = tf.reduce_mean(cost_i)
    return cost
```

```python
optimizer = Adam(learning_rate=0.1)
for i in range(2000):
    optimizer.minimize(compute_loss, var_list=[w1,w2,w3,  b1,b2,b3  ])
    print(i, compute_loss().numpy())
```

```python
def hxfn( xdata ):
    xd = tf.constant( xdata, tf.float32)
    z1 = tf.matmul( xd, w1) + b1  
    hx1 = tf.nn.softmax( z1 )
    
    z2 = tf.matmul(hx1,w2)+b2
    hx2 = tf.nn.softmax( z2 )
    
    z3 = tf.matmul(hx2, w3 ) + b3
    hx = tf.nn.softmax( z3 )
    
    return (hx.numpy()>0.5) + 0
```

```python
hxfn([[5.1,3.5,1.4,0.2]])
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score
print('정확도:',accuracy_score(y_data, hxfn(x_data)))
```

## 케라스

```python
dense1 = Dense( units=100, input_dim=4 ,activation="softmax") # 2x100, b:100
dense2 = Dense( units=3, input_dim=100 ,activation="softmax") # 100x1 b: 1
model = Sequential( [dense1, dense2])
model.compile( loss="categorical_crossentropy" ,optimizer=Adam(0.1), metrics=['acc'] )
h = model.fit( x_data, y_data, epochs=1000)
```
