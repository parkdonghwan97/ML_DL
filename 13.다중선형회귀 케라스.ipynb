```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.optimizers import Adam,Adagrad
import matplotlib.pyplot  as plt

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

import pandas as pd
plt.style.use(['seaborn-white'])
```

```python
df = pd.read_csv('data/test.csv', header=None)
df.columns=['q1','q2','mid','final']
df.head(3)
```

```python
x_data = df.iloc[:,:-1].values #특성데이터 추출
y_data = df.iloc[:, [-1]].values
```

```python
dense = Dense(units = 1, input_dim =3 ) # w : 3x1   b : 1   초기값
model = Sequential([dense])
model.compile(loss='mse',optimizer=Adam(0.1))
# n행 3열
h = model.fit(x_data, y_data, epochs=500) # x값을 행렬로 주어야함
```

```python
w,b = dense.get_weights()
w.shape,b.shape  # w = 3행 1열
```

```python
# q1 q2 mid  80 90 90 인 사람의 최종점수
model.predict([ 
    [80,90,90]
])
```
