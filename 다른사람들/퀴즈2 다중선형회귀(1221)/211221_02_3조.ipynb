보스턴 데이터 셋을 학습(텐서, 케라스) 하여 특성데이터의 0번째 행의 값의 주택중위가격을 예측하시오

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import seaborn as sns
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
df_boston = pd.DataFrame(boston['data'])
df_boston.columns = boston['feature_names']
df_boston['medv'] = boston['target']
df_boston
```

```python
x_data = df_boston.iloc[:,:-1].values #특성데이터 추출
y_data = df_boston.iloc[:, [-1]].values
```

```python
x_data.shape
```

```python
y_data.shape
```

```python
x = tf.constant( x_data, tf.float32)
y = tf.constant( y_data, tf.float32)
```

```python
w = tf.Variable( tf.random.uniform( [13,1] ) ) #[특성데이터갯수,라벨의갯수]
b = tf.Variable( tf.random.uniform( [1] ) ) #[라벨의갯수]
```

```python
def compute_loss():
    hx = tf.matmul( x, w) + b
    cost = tf.reduce_mean( (hx-y)**2)
    return cost
```

```python
optimizer = Adam(0.1)
for i in range(1000):
    optimizer.minimize( compute_loss, var_list=[w,b] )
    print(i, 'cost:', compute_loss().numpy() )
```

```python
def hxfn(xdata):
    xd = tf.constant( xdata ,tf.float32)
    hx = tf.matmul( xd, w) +b
    return hx.numpy()
```

```python
x_data[0]
```

```python
# 0행의 값
hxfn([x_data[0]])
```

```python
dense = Dense(units=1, input_dim=13)
model = Sequential([dense])
model.compile(loss='mse', optimizer=Adam(lr=0.1))
h=model.fit(x,y, epochs=1000)
```

```python
# 0행의 값
model.predict(x_data)[0]
```
