#### electric.csv 파일을 이용하여 텐서, 케라스로 각각 학습
    1) scatterplot + 예측선
    2) 전기생산량이 (3,4)인 경우 전기사용량을 예측
    3) 결정계수(mse)로 검증

```python
import tensorflow as tf
import numpy as np
import pandas as pd

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
df = pd.read_csv('data/electric.csv', index_col='Unnamed: 0')
df
```

```python
x_data = df['전기생산량'].values
y_data = df['전기사용량'].values
```

```python
plt.scatter(x_data, y_data)
plt.show()
```

#### 텐서 버전

```python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
```

```python
w = tf.Variable( tf.random.uniform([1]) ) #임의의 값 부여
b = tf.Variable( tf.random.uniform([1]) )
```

```python
def compute_cost():
    hx = w*x + b
    cost = tf.reduce_mean((hx-y)**2)
    return cost
```

```python
optimizer = Adam(learning_rate = 0.01)
electric_cost = []
for i in range(2000):
    optimizer.minimize(compute_cost, var_list=[w, b]) 
    c = compute_cost().numpy()
    electric_cost.append(c)
    print(i, 'cost = ', c)
```

```python
def hxFn(x_data):
    # xd = tf.constant( xdata)
    # hx = w*xd + b
    hx = w*x_data + b
    return hx.numpy()
```

```python
#전기생산량이 (3,4)인 경우 전기사용량
hxFn([3, 4])
```

```python
plt.plot(electric_cost)
plt.show()
```

```python
#예측선 그리기
plt.scatter(x_data, y_data)
plt.plot(x_data, hxFn(x_data), color='b')
plt.show()
```

#### 케라스 버전

```python
dense = Dense(units=1, input_dim=1)
model = Sequential([dense])
model.compile(loss='mse', optimizer = Adam(learning_rate=0.01))
h = model.fit(x_data, y_data, epochs = 1000)
```

```python
w, b = dense.get_weights()
```

```python
#전기생산량이 (3,4)인 경우 전기사용량
model.predict([3, 4])
```

```python
#텐서 : 0.02514846 / 케라스 : 0.0251
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_data, model.predict(x_data))
r2
```
