# Library Import

````python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
````

# Data

````python
df = pd.read_csv("data/electric.csv", index_col = 'Unnamed: 0')
df
````

````python
x_data = df['전기생산량'].values  
y_data = df['전기사용량'].values   
````

# Quiz (Tensor)

## scatter와 예측선

````python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
````

````python
w = tf.Variable( tf.random.uniform([1]) )
b = tf.Variable( tf.random.uniform([1]) )
````

````python
def compute_cost():
    hx = w*x + b
    cost = tf.reduce_mean( (hx-y)**2 )
    return cost
````

````python
optimizer = Adam(learning_rate=0.01)
hist_cost = []

for i in range(2000):
    optimizer.minimize( compute_cost, var_list=[w,b] )
    c = compute_cost().numpy()
    hist_cost.append(c)         
    print(i, "cost=", c)
````

````python
def hxFn( xdata):
    hx = w*xdata + b           
    return hx.numpy()
````

````python
plt.scatter( x_data, y_data)
plt.plot(x_data, hxFn(x_data), 'r')
plt.show()
````

## 전기생산량이 (3,4)인 경우 전기사용량 예측

````python
hxFn([3, 4])
````

## 결정계수

````python
from sklearn.metrics import r2_score
````

````python
y_true = y_data
y_pred = hxFn(x_data)

r2_score(y_true, y_pred)
````

# Quiz (Keras)

## scatter와 예측선

````python
import tensorflow as tf
import numpy as np
import pandas as pd


# 케라스에서 할 때 필요한 것
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense


from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
````

````python
dense = Dense( units=1, input_dim=1)                  
    
model = Sequential( [dense] )                       

model.compile( loss='mse', optimizer=Adam(0.1) )   

h = model.fit( x_data, y_data, epochs=500)         
````

````python
plt.scatter( x_data, y_data)
plt.plot(x_data, model.predict(x_data) , 'r')
plt.show()
````

## 전기생산량이 (3,4)인 경우 전기사용량 예측

````python
model.predict( [3, 4] )
````

## 결정계수

````python
from sklearn.metrics import r2_score
````

````python
y_true = y_data
y_pred = model.predict(x_data)

r2_score(y_true, y_pred)
````
