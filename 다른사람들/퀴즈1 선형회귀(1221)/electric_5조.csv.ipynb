# 텐서

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.optimizers import Adam 
import matplotlib.pyplot as plt
import pandas as pd
from keras import metrics

from sklearn.metrics import r2_score
```

```python
df= pd.read_csv('data/electric.csv', index_col='Unnamed: 0')
df.sample(5)
```

```python
x=df['전기생산량']
y=df['전기사용량']

x=tf.constant(x, tf.float32)
y=tf.constant(y, tf.float32)
```

```python
plt.scatter(x,y)
```

```python
w=tf.Variable(tf.random.uniform([1]))
b=tf.Variable(tf.random.uniform([1]))
```

```python
def compute_cost():
    hx=w*x+b
    cost=tf.reduce_mean((hx-y)**2)
    return cost
```

```python
optimizer=Adam(learning_rate=0.01)
hist_cost=[]

for i in range(2000):
    optimizer.minimize(compute_cost ,var_list=[w,b] )
    c=compute_cost().numpy()
    hist_cost.append(c)
    print(f'{i+1} cost={c}')
```

```python
plt.plot(hist_cost)
```

```python
def hxFn(x):
    hx=w*x+b 
    return hx.numpy()
```

```python
hxFn([3,4])
```

```python
plt.scatter(x, y)
plt.plot(x,hxFn(x), 'r--')
```

```python
# rmse
r2_score(y, hxFn(x))
```

# 케라스

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
dense=Dense(units=1, input_dim=1) #w,b 초기값
model=Sequential([dense])
model.compile(loss='mse', optimizer=Adam(0.01))
h=model.fit(x, y, epochs=500)
```

```python
w,b=dense.get_weights()
```

```python
model.predict([3,4])
```

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x), 'r--')
```

```python
#결정계수
```

```python
r2_score(y, model.predict(x))
```
