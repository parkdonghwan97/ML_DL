## scatter과 예측선

```python
x_data = torch.FloatTensor(exdf.전기생산량)
y_data = torch.FloatTensor(exdf.전기사용량)
```

```python
w = torch.tensor(10.0, requires_grad = True)
b = torch.tensor(10.0, requires_grad = True)
```

```python
optimizer = Adam([w, b], lr = 0.1)
```

```python
for i in range(3000):
    hx = w * x_data + b
    cost = torch.mean((hx - y_data)**2)
    optimizer.zero_grad()# grad 값을 zeos해줌.
    cost.backward() # 미분 함수. w, b에 대한 미분 수행
    optimizer.step() # w 미분값, b에 대한 미분 값
    print(i, 'step cost = ', cost.item())
```

```python
w, b
```

```python
def hxfn(xdata):
    # 기존에는 tensor node로 만들어줘야만 가능. 하지만 지금은 list로 데이터를 받는게 가능.
    xd = torch.FloatTensor(xdata)
    hx = w * xd + b
    # w가 tensor 객체이기에 연산함.
    return hx.detach().numpy()
```

```python
plt.scatter( x_data, y_data)
plt.plot(x_data, hxfn(x_data), color='r')
plt.show()
```

## mse(결정계수?? r^2)로 검증하시오.

- mse가 cost function이라서 2999 step cost =  0.02514859102666378

$R^{2} = 1 - \frac{\sum_{i}(y_i-\hat{y_i})^2}{\sum_i(y_i-\overline{y})^2}$

```python
y_pred = hxfn(x_data)
```

```python
def r2_loss(output, target):
    target_mean = torch.mean(target)
    ss_tot = torch.sum((target - target_mean) ** 2)
    ss_res = torch.sum((target - output) ** 2)
    r2 = 1 - ss_res / ss_tot
    return r2
```

```python
r2_loss(y_pred, y_data)
```
