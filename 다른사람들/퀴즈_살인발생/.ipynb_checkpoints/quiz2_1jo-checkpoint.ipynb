```python
x_data = df['살인 발생'].values
y_data = df['폭력 발생'].values

x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
```

```python
w = tf.Variable(tf.random.uniform([1]))
b = tf.Variable(tf.random.uniform([1]))
```

```python
def compute_loss():
    hx = w*x + b
    cost = tf.reduce_mean((hx-y)**2)
    return cost
```

```python
optimizer = Adam(0.5)
hist_cost = []
for i in range(2000):
    optimizer.minimize(compute_loss, var_list = [w,b])
    c = compute_loss().numpy()
    hist_cost.append(c)
    print( i, "cost:", c)
```

```python
plt.plot(hist_cost)
```

```python
def hxFn(xData):
    #xd = tf.constant(xdata) <-- 원래는
    hx = w*xData + b
    return hx.numpy()
```

```python
plt.scatter( x_data, y_data)
plt.plot(x_data, hxFn(x_data),'r--')
```

```python
plt.plot(y_data)
plt.plot(hxFn(x_data),'r--')
```

```python
hxFn(5), df[df['살인 발생'] == 5]['폭력 발생']
```

```python
dense = Dense( units=1, input_dim = 1) # w,b 초기값...
model = Sequential([dense])
model.compile(loss='mse', optimizer = Adam(0.5))
h = model.fit(x_data, y_data, epochs=2000)
```
