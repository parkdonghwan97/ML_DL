## 3번) Tensor version

````python
x_data = df_dummi.iloc[:, [0,3,4,5,6,7,8]]
y_data = pd.get_dummies(df['Survived'])
````

### Tensor node

````python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
````

### 매개변수 초기화

````python
#xavier, he ==> ( 최적의 초기화 알고리즘 )
initX = tf.initializers.GlorotUniform()
w = tf.Variable( initX(shape=[ x_data.shape[1], y_data.shape[1]] ) )    
b = tf.Variable( initX(shape=[ y_data.shape[1]] ) )               
````

### Cost Func.

````python
def compute_loss():
    z = tf.matmul(x,w) + b     
#   hx = tf.nn.sigmoid( z )    
    cost_i = tf.nn.sigmoid_cross_entropy_with_logits( labels=y, logits=z )
    cost = tf.reduce_mean( cost_i )
    return cost
````

### Optimization

````python
optimizer = Adam( 0.1 )
hist_cost = []

for i in range(1000):
    optimizer.minimize( compute_loss, var_list=[w,b] )
    c = compute_loss().numpy()
    hist_cost.append(c)
    print(i, 'cost:', c)
````

### Prediction

````python
def hxfn(xdata):
    xd = tf.constant( xdata, tf.float32)
    z = tf.matmul( xd, w ) + b
    hx = tf.nn.sigmoid( z )
    return hx.numpy()         
````
