```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.optimizers import Adam,Adagrad
import matplotlib.pyplot  as plt
import pandas as pd
plt.style.use(['seaborn-white'])
```

```python
df = pd.read_csv('data/cars.csv',index_col='Unnamed: 0') 
df.head(3)
```

```python
x_data = df['speed'].values
y_data = df['dist'].values
```

```python
plt.scatter(x_data,y_data)
```

```python
x = tf.constant(x_data,tf.float32)      # 자동차 속도
y = tf.constant(y_data,tf.float32)      # 제동 거리
```

```python
w = tf.Variable(tf.random.uniform([1]))
b = tf.Variable(tf.random.uniform([1]))
```

```python
def compute_cost():
    hx = w * x  + b
    cost = tf.reduce_mean( (hx - y)**2 )
    return cost  
```

```python
optimizer = Adam(learning_rate=0.01)
hist_cost = []
for i in range(2000):
    optimizer.minimize(compute_cost, var_list=[w,b])
    c = compute_cost().numpy()
    hist_cost.append( c )
    print(i,'cost= ', c)
```
